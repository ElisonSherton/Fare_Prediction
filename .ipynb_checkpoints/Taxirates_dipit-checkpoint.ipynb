```python
#distance = R*(sqrt((dLat)**2 + (cos(meanLat)*dLong)**2))
from math import pi,sqrt,cos
def distance(lat1,lat2,long1,long2):
    lat1 = lat1*(pi/180)
    lat2 = lat2*(pi/180)
    long1 = long1*(pi/180)
    long2 = long2*(pi/180)
    meanLat =(lat1+lat2)/2
    dLat = (lat2-lat1)
    dLong = (long2-long1)
    R = 6371.009
    distance = R*(sqrt((dLat)**2 + (cos(meanLat)*dLong)**2))
    return distance
```

```python
for col in ['vendor_id', 'new_user','mta_tax','passenger_count', 'rate_code', 'store_and_fwd_flag', 'payment_type',
           'surcharge']:
    new_rate24[col]=new_rate24[col].astype('category')
```

```python
cat_cols = ['vendor_id', 'new_user','mta_tax','passenger_count', 'rate_code', 'store_and_fwd_flag', 'payment_type',
           'surcharge']
num_cols = ['tolls_amount', 'tip_amount','hour', 'distance']
```

```python
X = new_rate24.drop(['fare_amount'],axis=1)
y = new_rate24['fare_amount']
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2,random_state = 0)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

#### Imputing the missing values

```python
df_cat_train = X_train[cat_cols]
df_cat_test = X_test[cat_cols]
df_num_train = X_train[num_cols]
df_num_test = X_test[num_cols]
```

```python
df_cat_train.mode()
```

```python
#Impute on train
df_cat_train.fillna(df_cat_train.mode().iloc[0],inplace=True)
# Impute on test
df_cat_test.fillna(df_cat_train.mode().iloc[0],inplace=True)
```

```python
df_cat_train.isnull().sum()
```

```python
df_cat_train.head()
```

#### Imputing the numerical values

```python
df_num_train.fillna(df_num_train.median(),inplace=True)
df_num_test.fillna(df_num_train.median(),inplace=True)
```

```python
df_num_train.isnull().sum()
```

```python
df_num_test.isnull().sum()
```

```python
X_train = pd.concat((df_cat_train,df_num_train),axis=1)
X_test = pd.concat((df_cat_test,df_num_test),axis=1)
```

```python
print(X_train.head())
```

```python
#data = pd.get_dummies(columns=categorical_Attributes, data=data, 
                      #prefix=categorical_Attributes, prefix_sep="_", drop_first=True)
X_train = pd.get_dummies(columns = cat_cols,data=X_train,prefix = cat_cols,prefix_sep="_")
X_test = pd.get_dummies(columns = cat_cols,data=X_test,prefix = cat_cols,prefix_sep='_')
```
